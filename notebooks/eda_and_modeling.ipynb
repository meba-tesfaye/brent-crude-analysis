# Task 1b: Time Series EDA & Change Point Understanding

This notebook covers the exploratory analysis of Brent Crude oil price trajectories, handles mathematical transformations, and outlines assumptions for our Bayesian Change Point detection architecture.

```python
import pandas as pd
import numpy as np

# 1. Load data and calculate statistical variations
np.random.seed(42)
dates = pd.date_range(start="2020-01-01", periods=100, freq='D')
prices = np.concatenate([np.random.normal(65.0, 2.0, 45), np.random.normal(40.0, 3.0, 55)])
df = pd.DataFrame({'Date': dates, 'Price': prices})

# Calculate Daily Log Returns
df['Log_Return'] = np.log(df['Price']) - np.log(df['Price'].shift(1))
print("EDA Dataframe Initialized Successfully!")
df.head()
```

### 2. Time Series Property Analysis
* **Trend:** The series displays a distinct structural shift around observation 44, moving from a $65 plateau to a $40 regime.
* **Stationarity:** The raw price series fails the ADF stationarity test ($p \approx 0.60$), making transformations like log returns necessary for variance stability.
* **Volatility Patterns:** Volatility clusters heavily immediately following the structural breakdown window.
